(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements.

```
# write your answer here
from array import *;
my_array = array('i', [1,2,3,4,5,1,2,3,4,23,32,45,32,21,23])
def Repeat(x): 
    _size = len(x) 
    repeated = [] 
    for i in range(_size): 
        k = i + 1
        for j in range(k, _size): 
            if x[i] == x[j] and x[i] not in repeated: 
                repeated.append(x[i]) 
    if not repeated:
      print('No Duplicates!')
      return -1;
    else:
      print('Duplicates:')
      return repeated;
print (Repeat(my_array)) 

```

(2) Write a Python program to select all the Sundays of a specified year.



```
# write your answer here
from datetime import date, timedelta
def all_sundays(year):
       dt = date(year, 1, 1)     
       dt += timedelta(days = 6 - dt.weekday())  
       while dt.year == year:
          yield dt
          dt += timedelta(days = 7)
input_year=int(input('Enter the Year:'))
for s in all_sundays(input_year):
   print(s)
```
